**HMRC-COMPLIANT UK SME REGIONAL TURNOVER ANALYSIS**

This analysis examines the regional turnover performance of UK businesses based on HMRC's official small business classification thresholds (Micro-entity: ≤£0.632M, Small Business: ≤£10.2M, SME/Medium: >£10.2M). The dataset covers the 2018 monthly financial records of 1 SCX Special Projects, headquartered in Sheffield, integrating profit margin, tax liability, and employee count variables.

**HMRC Compliance Note: All business classifications are aligned with the Companies Act 2006 and HMRC SME definition criteria.**

```python
df_final["profit_margin"] = (df_final["pre-tax profit"] / df_final["turnover"].replace(0, np.nan)) * 100
df_final["profit_margin"] = df_final["profit_margin"].fillna(0)

def classify_hmrc(row):
    if row['turnover'] <= 0.632: return 'Micro-entity'
    elif row['turnover'] <= 10.2: return 'Small Business'
    else: return 'SME / Medium'

df_final['business_size'] = df_final.apply(classify_hmrc, axis=1)
df_final.head()
```

```python
#BOXPLOT GRAPH
plt.figure(figsize=(6,4))
sns.boxplot(x=df_final["turnover"])
plt.title("Outliers and Extreme Value",fontsize=20)
plt.show()
```

```python
# IQR-based outlier flagging (better than visual boxplot alone)
Q1 = df_final["turnover"].quantile(0.25)
Q3 = df_final["turnover"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_final["turnover_outlier_flag"] = (
    (df_final["turnover"] < lower_bound) | 
    (df_final["turnover"] > upper_bound)
).astype(int)

outlier_count = df_final["turnover_outlier_flag"].sum()
outlier_pct   = df_final["turnover_outlier_flag"].mean() * 100

print(f"IQR Bounds: Lower = £{lower_bound:,.2f}M  |  Upper = £{upper_bound:,.2f}M")
print(f"Outliers detected: {outlier_count} companies ({outlier_pct:.1f}% of dataset)")
print("\nTop 10 outlier companies:")
print(df_final[df_final["turnover_outlier_flag"] == 1]
      [["smes","location","activity","turnover"]]
      .sort_values("turnover", ascending=False).head(10))
```

```python
# Z-score outlier analysis (complementary method)
from scipy import stats
z_scores = np.abs(stats.zscore(df_final["turnover"].dropna()))
extreme_outliers = (z_scores > 3).sum()
print(f"Companies with Z-score > 3 (extreme outliers): {extreme_outliers}")
print(f"\nStrategic Decision: Outliers RETAINED — they represent real market"
      " leaders and are not data errors.")
print("GetClarity Insight: These high-turnover businesses represent the 'Medium'"
      " tier that GetClarity Pro and Business plans are designed to serve.")
```

**Objective:** Assessing the validity of extreme observations and their impact on analysis.

- **Threshold Identification:** Companies exceeding the **£100,000** turnover mark appear statistically as "outliers," yet they show a consistent distribution among themselves. A hyper-niche group exceeding **£200,000** represents the extreme upper tier of the dataset.
- **Data Integrity Policy:** > **Strategic Decision:** These observations are a true reflection of the SME landscape and do not result from artificial data entry errors. To preserve the **natural flow and authenticity** of the market structure, no suppression or cleaning will be applied to these extreme values.

```python
# GetClarity Insight: Sectoral Profit vs HMRC Risk Flag
sector_summary = df_final.groupby("activity").agg(
    avg_margin=("profit_margin", "mean"),
    margin_std=("profit_margin", "std"),
    count=("smes", "count")
).round(2).sort_values("avg_margin", ascending=False)

sector_summary["hmrc_risk_flag"] = (
    (sector_summary["avg_margin"] < 0) | 
    (sector_summary["margin_std"] > 20)
).map({True: "⚠ HIGH RISK", False: "✓ OK"})

print("=== SECTOR PROFIT MARGIN + HMRC RISK FLAGS ===")
print(sector_summary.to_string())
```

```python
# GetClarity AI Copilot signal: tax efficiency ratio
df_final["tax_efficiency_ratio"] = (
    df_final["tax"] / df_final["turnover"].replace(0, np.nan)
).fillna(0).round(4)

efficiency_by_sector = (
    df_final.groupby("activity")["tax_efficiency_ratio"].mean()
    .sort_values(ascending=False)
    .round(4)
)
print("=== TAX EFFICIENCY RATIO BY SECTOR ===")
print("(Higher = paying more tax per £ of turnover)")
print(efficiency_by_sector)
```

```python
# GetClarity: Revenue growth rate (year-on-year proxy)
# We use earliest vs latest date per SME as a growth signal
first_last = df_final.groupby("smes")["turnover"].agg(["first","last"])
first_last["growth_pct"] = ((first_last["last"] - first_last["first"]) / 
                             first_last["first"].replace(0, np.nan) * 100).round(2)

print("=== SME REVENUE GROWTH (first vs latest observation) ===")
print(first_last.describe())

growing = (first_last["growth_pct"] > 0).sum()
declining = (first_last["growth_pct"] < 0).sum()
print(f"\nGrowing businesses : {growing}")
print(f"Declining businesses: {declining}")
print(f"GetClarity Target  : {declining} businesses need proactive AI cash-flow alerts.")
```

---

## <b style="font-size:24px; color:#21618C;">7.1 Objectives & Legal Framework</b>

**Objective:** To compare turnover and tax burden according to business size by labeling the data according to HMRC standards.

- **Legal Compliance:** The threshold values used in this analysis (**£0.632 million** and **£10.2 million**) are official legal limits set by the **UK Companies Act 2006** and **HMRC**. 
- **Classification Impact:** Whether a business is considered "Micro", "Small" or "Medium" in the UK depends on the turnover and the number of employees. These limits determine the tax obligations and financial reporting standards of businesses.

---

## <b style="font-size:24px; color:#21618C;">7.2 Threshold Definitions</b>

- **£0.632 Million (Micro-entity):** According to HMRC, a business is considered micro-sized (similar to a sole proprietorship/sole trader structure) if its turnover does not exceed £632,000 per year.
- **£10.2 Million (Small Business):** Businesses with an annual turnover below £10.2 Million and fewer than 50 employees are considered in the "Small Business" category.

---

## <b style="font-size:24px; color:#21618C;">7.3 Analytical Accuracy & Data Representation</b>

- **Segmentation Strategy:** Segmenting businesses in the dataset according to these official limits allows us to compare the financial behaviors, tax payments, and cash flow cycles of the **"Sole Trader"** and **"SME"** groups in a manner compatible with the UK economy.

> **Technical Note:** Since the turnover figures in our verification set are in millions of pounds, the legal limit of £632,000 is represented as **0.632** in the code, and the value of £10,200,000 is represented as **10.2**.


```python
conditions=[(df_final["turnover"]<0.632)&(df_final["no of employee"]<=10),(df_final["turnover"]<=10.2)&(df_final["no of employee"]<=50),(df_final["turnover"]>10.2)]
choices=["sole_trader","small_business","medium"]
df_final["business_type"]=np.select(conditions,choices,default="medium")
```

```python
#average income/tax statement profile
df_final.groupby("business_type")[["turnover","tax"]].mean().reset_index()
```

## <b style="font-size:24px; color:#21618C;">3. Analysis of HMRC Allowable Expenses and Profit Structure</b>

The success of a business is measured not only by its turnover, but also by how much of this turnover goes to operational expenses. According to HMRC standards, business owners can only deduct from tax expenses that are **"wholly and exclusively for business"** (directly related to the business). 

In this analysis, the financial structure of a typical SME in the UK has been simulated:

* **Operating Costs:** Covers items such as rent, office expenses and invoices. It usually makes up the largest part of the budget.
* **Staff Wages (Personnel Expenses):** This group includes salaries, insurance premiums (National Insurance), and pension contributions.
* **Allowable Expenses:** Represents tax-deductible expenses approved by HMRC, such as advertising, travel, and stock purchases.
* **Tax and Net Profit:** Shows the actual profit margin remaining with the business after taxes.

> **Methodological Note:** Since the raw dataset does not contain specific expense columns, this distribution is based on general HMRC SME benchmarks to provide a strategic financial overview.

## <b style="font-size:24px; color:#21618C;">9.2 Cash Flow Cycles and Monthly Revenue Trends</b>

**Objective:** Identifying high-season peaks and low-season shortages to ensure operational liquidity.

- **Vital Metric:** Understanding cash flow is vital for any SME. While turnover shows total sales, the timing of that income determines if a business can pay its bills on time. 
- **Trend Analysis:** In this section, we analyze the monthly trend of revenue to identify high-season peaks and low-season shortages.

```python
df_final["month"]=df_final["date"].dt.month
```

```python
monthly_revenue=df_final.groupby("month")["turnover"].mean().reset_index()
```

```python
CLARITY_TEAL = "#1D9E75"

plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly_revenue, x="month", y="turnover",
             marker="o", color=CLARITY_TEAL, linewidth=2.5)
plt.fill_between(monthly_revenue["month"], monthly_revenue["turnover"],
                 alpha=0.15, color=CLARITY_TEAL)
plt.title("Monthly Revenue Trend — Cash Flow Cycle Analysis",
          fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Month", fontsize=11, labelpad=15)
plt.ylabel("Average Turnover (£ Millions)", fontsize=11, labelpad=15)
plt.xticks(range(1, 13),
           ["Jan","Feb","Mar","Apr","May","Jun",
            "Jul","Aug","Sep","Oct","Nov","Dec"])
plt.grid(True, linestyle="--", alpha=0.5)

# Annotate peak and trough
peak_month = monthly_revenue.loc[monthly_revenue["turnover"].idxmax()]
trough_month = monthly_revenue.loc[monthly_revenue["turnover"].idxmin()]
plt.annotate(
    f"Peak\n£{peak_month["turnover"]:.0f}M",
    xy=(peak_month["month"], peak_month["turnover"]),
    xytext=(peak_month["month"] + 0.4, peak_month["turnover"] * 1.03),
    fontsize=9, color="#1D9E75", fontweight="bold"
)
plt.annotate(
    f"Trough\n£{trough_month["turnover"]:.0f}M",
    xy=(trough_month["month"], trough_month["turnover"]),
    xytext=(trough_month["month"] + 0.4, trough_month["turnover"] * 0.97),
    fontsize=9, color="#922B21", fontweight="bold"
)
plt.tight_layout()
plt.savefig("cash_flow_cycle.png", dpi=150, bbox_inches="tight")
plt.show()
```
